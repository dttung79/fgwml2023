```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
#import matplotlib.pyplot as plt
#import tensorflow as tf
from tensorflow import keras
```

```python
data = np.loadtxt("wine.csv", delimiter=",", skiprows=1)
inputs = data[:, :11]
targets = data[:, 11]
min_t = min(targets)
max_t = max(targets)
# scale targets to 0-1 range
#targets = (targets - min_t) / (max_t - min_t)

inputs_train = inputs[:4500]
targets_train = targets[:4500]
inputs_test = inputs[4500:]
targets_test = targets[4500:]

input_size = 11
output_size = 1
```

```python
# Build a neural network
# Setup some hyperparameters
WIDTH = 50
LEARNING_RATE = 0.00001
# Create a sequential model
model = Sequential()
# Add input layer
model.add(Dense(WIDTH, activation="elu", input_dim=input_size, kernel_initializer="uniform"))
# Add hidden layer
model.add(Dense(WIDTH, activation="relu", kernel_initializer="uniform"))
model.add(Dense(WIDTH, activation="selu", kernel_initializer="uniform"))
model.add(Dense(WIDTH, activation="selu", kernel_initializer="uniform"))
model.add(Dense(WIDTH, activation="selu", kernel_initializer="uniform"))
# Add output layer
model.add(Dense(1, activation="linear", kernel_initializer="uniform"))

# Compile model
model.compile(loss='mse', optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE))
model.fit(inputs_train, targets_train, epochs=1000, verbose=2)
```

```python
output_predict = model.predict_on_batch(inputs_test).round(0)
# rescale
#output_predict = output_predict * (max_t - min_t) + min_t
print(np.reshape(output_predict[:20], (20,)))
```

```python
targets_round = targets_test.round(3)
# rescale
#targets_round = targets_round * (max_t - min_t) + min_t
print(targets_round[:20])
```

```python
errors = np.abs(output_predict - targets_round)
print(errors)
```

```python
print('Euclidean distance:', np.linalg.norm(output_predict - targets_round))
```
