```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.linspace(-100, 100, 200)
f = lambda x: 5*(x**2) + 3*x - 4
t = f(w)
```

```python
nloops = 50
x = [0]*nloops
y = [0]*nloops
x[0] = 100
y[0] = f(x[0])
eta = 0.03
f1 = lambda x: 10*x + 3
for i in range(1, nloops):
    x[i] = x[i-1] - eta * f1(x[i-1])
    y[i] = f(x[i])

plt.plot(w, t, 'g-')
plt.plot(x, y, 'ro')
plt.show()
```
